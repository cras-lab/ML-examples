임의의 2차원 배열을 생성한 후 PCA 수행<BR>
먼저 필요한 모듈을 임포트 한다.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

임의의 2차원 데이터를 생성한다. 여기서는 다변량 정규분포 데이터를 생성한다.
image.png

```python
# 2차원 데이터 생성
np.random.seed(0)
mean = [0, 0]     # 평균 벡터의 지정
cov = [[1, 0.8], [0.8, 1]]  # 공분산 행렬의 지정
X = np.random.multivariate_normal(mean, cov, 100)  # 2차원 데이터 100개 생성
```

PCA 모델을 생성한다.

```python
# PCA 모델 생성
pca = PCA(n_components=1)
```

PCA를 거친 데이터를 평균이 0이고 분산이 1이 되도록 정규화 한다.

```python
# 데이터 변환
X_transformed = pca.fit_transform(X)   # 정규화, 평균이 0, 분산이 1
```

원시데이터와 변환된 데이터를 같이 시각화 한다.

```python
# 랜덤 데이터 100개 시각화
plt.figure(figsize=(6, 6))
plt.scatter( X[:,0], X[:,1] )


# 변환된 데이터 시각화
plt.scatter(X_transformed, np.zeros_like(X_transformed))

plt.xlabel('Principal Component')
plt.ylabel('')
plt.title('Data Transformed by PCA')
plt.grid(True)
plt.show()
```
