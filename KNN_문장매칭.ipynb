LLM을 쓰지 않고, 오직 KNN 유사도로 찾는 예제<BR>
필요한 모듈을 임포트 한다.

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sentence_transformers import SentenceTransformer
```

예제 문장을 입력

```python
sentences = [
    "강의 속도가 너무 빨라서 따라가기 어려웠습니다.",
    "조금 천천히 설명해 주시면 좋겠습니다.",
    "설명이 전반적으로 명확하고 이해하기 쉬웠습니다.",
    "예제가 많아서 개념을 이해하는 데 큰 도움이 됐습니다.",
    "과제가 생각보다 어려웠지만 많이 배웠습니다.",
    "과제 난이도가 너무 높아서 부담이 컸습니다.",
    "과제 피드백을 더 자세히 받고 싶습니다."
]
```

문장 임베딩을 위한 객체를 생성한다.

```python
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

문장을 임베딩한다.

```python
doc_embeddings = model.encode(sentences, show_progress_bar=False)
```

코사인유사도를 사용해 k=3인 KNN을 수행한다.

```python
knn = NearestNeighbors(n_neighbors=3, metric="cosine")
knn.fit(doc_embeddings)
```

루프를 돌며, 질문에 답변한다.<BR>
KNN은 단순 문장 매칭이므로, 긍정 부정 등의 감정 매핑은 약하다.<BR>

```python
while True:
    question = input("질문: ").strip()
    if question in ["exit", "quit", "종료"]:
        break

    query_embedding = model.encode([question], show_progress_bar=False)
    distances, indices = knn.kneighbors(query_embedding)

    print("\n[가장 관련 있는 문장]")
    for i, idx in enumerate(indices[0]):
        print(f"{i+1}. {sentences[idx]}  (distance={distances[0][i]:.4f})")
    print()
```
